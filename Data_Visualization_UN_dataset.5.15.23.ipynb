```python
import matplotlib.pyplot as mpl
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#Dataset from https://www.kaggle.com/datasets/ammaraahmad/immigration-to-canada?resource=download
df_canada=pd.read_csv('canadian_immegration_data.csv')
#Make 'country index column'
df_canada.set_index('Country',inplace = True)

df_canada.head()
```

## **Line Plots**

For continuous data over a period of time

In 2010, Haiti suffered a catastrophic magnitude 7.0 earthquake. The quake caused widespread devastation and loss of life and about **three million people** were affected by this natural disaster. As part of Canada's humanitarian effort, the Government of Canada stepped up its effort in accepting refugees from Haiti. Visualize this effort using a Line plot:

```python
years = list(map(str, range(1980, 2014)))
```

```python
#Get the country Haiti and the years 1980-2013
haiti=df_canada.loc['Haiti', years]

#Change years from str to int
haiti.index = haiti.index.map(int)

#Create line graph
haiti.plot(kind='line')
mpl.title('Immigration From Haiti')
mpl.xlabel('Years')
mpl.ylabel('Number of Immigrants')

#Add text when earthquake struck (2010)
mpl.text(2000, 6000, '2010 Earthquake')
mpl.show()
```

Compare the trend of top 5 countries that contributed the most to immigration to Canada.

```python
#Sort the Total column to get the top 5 countries
df_canada.sort_values(by='Total', ascending=False, axis=0, inplace=True) 

# Get the top 5 entries
df_top5 = df_canada.head(5)
df_top5
```

```python
 #Transpose the dataframe
df_top5 = df_top5[years].transpose() 
df_top5
```

```python
#Plot the dataframe
df_top5.index = df_top5.index.map(int) # Change the index values to integer 
df_top5.plot(kind='line', figsize=(14, 8))

mpl.title('Immigration Trend of Top 5 Countries')
mpl.ylabel('Number of Immigrants')
mpl.xlabel('Years')
mpl.legend(frameon=False) #Remove legend border

mpl.show()
```

## Area Plots
Depicts accumulated totals with numbers/percentages over time

```python
#Plot the dataframe
df_top5.index = df_top5.index.map(int) # Change the index values to integer 
df_top5.plot(kind='area', figsize=(14, 8)) #To make unstacked plot, stacked=False
                                           #To increase transparency, alpha=(# put in value 0-1)

mpl.title('Immigration Trend of Top 5 Countries')
mpl.ylabel('Number of Immigrants')
mpl.xlabel('Years')
mpl.legend(frameon=False) #Remove legend border

mpl.show()
```

```python
#Object oriented method, use an Axes instance of your current plot and store it in a variable (eg. ax)
ax = df_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('Immigration Trend of Top 5 Countries', fontsize=20)
ax.set_ylabel('Number of Immigrants', fontsize=14)
ax.set_xlabel('Years', fontsize=14)
ax.legend(frameon=False) #Remove legend border
```

**Create a stacked area plot of the 5 countries that contributed the least to immigration to Canada from 1980 to 2013. Use a transparency value of 0.45.**

```python
df_canada.sort_values(by='Total', axis=0, inplace=True) #Default is ascending order
df_low5=df_canada.head()
df_low5
```

```python
df_low5=df_low5[years].transpose()
```

```python
#Plot the dataframe
df_low5.index = df_top5.index.map(int) # Change the index values to integer 
df_low5.plot(kind='area', alpha=0.45, figsize=(12, 8))

mpl.title('Immigration Trend of Bottom 5 Countries')
mpl.ylabel('Number of Immigrants')
mpl.xlabel('Years')
mpl.legend(frameon=False, loc='upper left') #Remove legend border, move to upper left

mpl.show()
```

```python
#Using object-oriented method
ax = df_low5.plot(kind='area', alpha=0.45, stacked=False, figsize=(20, 10))
    
ax.set_title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
ax.legend(frameon=False)
```

## Histograms
Represents a variables frequency distribution.

**What is the frequency distribution of the number (population) of new immigrants from the various countries to Canada in 2013?**

```python
df_canada['2010'].plot(kind='hist')

mpl.title('Immigration From 195 Countries in 2010')
mpl.xlabel('Number of Immigrants')
mpl.ylabel('Number of Countries')
```

```python
count, bin_edges=np.histogram(df_canada['2010']) #Default computes 10 bins of equal width
print('Frequency Count:', count, 'Bin Ranges:', bin_edges)
```

```python
df_canada['2010'].plot(kind='hist', xticks=bin_edges, edgecolor='black') #xticks contain list of bin sizes

mpl.title('Immigration From 195 Countries in 2010')
mpl.xlabel('Number of Immigrants')
mpl.ylabel('Number of Countries')
```

**What is the immigration distribution for Denmark, Norway, and Sweden for years 1980 - 2013?**

```python
df_scndvia=df_canada.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
```

```python
count, bin_edges=np.histogram(df_scndvia)

df_scndvia.plot(kind='hist', xticks=bin_edges, edgecolor='black') 

mpl.title('Immigration From Scandinavia to Canada from 1980-2013')
mpl.xlabel('Number of Immigrants')
mpl.ylabel('Number of Years')
```

```python
#To make the graph more aesthetically pleasing:
count, bin_edges = np.histogram(df_scndvia, 15) #15 bins of equal width
xmin = bin_edges[0] - 10   #  first bin value is 31.0, adding buffer of 10 for aesthetic purposes 
xmax = bin_edges[-1] + 10  #  last bin value is 308.0, adding buffer of 10 for aesthetic purposes

#Create histogram
df_scndvia.plot(kind ='hist', 
          figsize=(10, 6),
          bins=15, #Increase bin size to 15
          alpha=0.6, #Transparency to 60%
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
          stacked=True, #No plot overlap
          xlim=(xmin,xmax) #Remove the extra gap on the edges of the plot
         )

mpl.title('Immigration from Scandinavia (Denmark, Norway, and Sweden) from 1980 - 2013')
mpl.ylabel('Number of Years')
mpl.xlabel('Number of Immigrants')

```

## Bar Charts
Compare variable values at a specific point in time.

**Create a bar chart to visualize immigration from Iceland to Canada from 1980-2013.**

```python
df_iceland=df_canada.loc['Iceland', years]
df_iceland.head()
```

```python
df_iceland.plot(kind='bar', figsize=(10,8))
   
mpl.title('Icelandic immigrants to Canada from 1980 to 2013')
mpl.xlabel('Year')
mpl.ylabel('Number of Immigrants')
mpl.show()
```

**Add an arrow and text to denote the 2008 2011 Icelandic Financial Crisis**

```python
df_iceland.plot(kind='bar', figsize=(10, 6), rot=90)  # rotate the xticks(labeled points on x-axis) by 90 degrees

mpl.xlabel('Year')
mpl.ylabel('Number of Immigrants')
mpl.title('Icelandic Immigrants to Canada from 1980 to 2013')

#Annotate arrow
mpl.annotate('',  #leave blank for no text
             xy=(32, 70),  #Head of the arrow at point (year:2012 , pop:70)
             xytext=(28, 20),  #Base of the arrow at point (year:2008 , pop:20)
             xycoords='data',  #Use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', #Standard arrow style
                             connectionstyle='arc3', #Straight line connection type
                             color='r', lw=2)
             )
#Annotate Text
mpl.annotate('2008-2011 Financial Crisis',  #Text
             xy=(28, 30),  #Text starts at point (year 2008 , pop 30)
             rotation=72.5,  #Rotate to match arrow
             va='bottom',  #Text is vertically 'bottom' aligned
             ha='left',  #Text is horizontally 'left' algned.
             )

mpl.show()
```

**Using the scripting later and the df_can dataset, create a horizontal bar plot showing the total number of immigrants to Canada from the top 15 countries, for the period 1980 - 2013. Label each country with the total immigrant count.**

```python
df_canada.sort_values(by='Total', ascending=True, axis=0, inplace=True) 

# Get the top 15 entries
df_top15 = df_canada['Total'].tail(15)
df_top15
```

```python
df_top15.plot(kind='barh', figsize=(12,12), color='b')
mpl.xlabel('Number of Immigrants')
mpl.title('Top 15 Countries Contributing to the Immigration to Canada between 1980 - 2013')

#Annotate value labels to each country
for index, value in enumerate(df_top15): 
    label = format(int(value), ',') #Format int with commas
    
#Place text at the end of bar (subtracting 47000 from x, and 0.1 from y to make it fit within the bar)
mpl.annotate(label, xy=(value - 47000, index - 0.10), color='white')

mpl.show()
```
